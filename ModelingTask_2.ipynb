```python
import sys
import math
import matplotlib.pyplot as plt
import numpy as np


#Initialize global variables
v = 0.5                             # 0.5m/s propagation velocity
A_init = 1                          # Amplitude = 1
wavelength_init = 50                # Iniital wavelength = 50m
t_total = 200                       # 200s total time
min_x = 0
max_x = 50
L_x = 50                            # Domain Size 50 m
num_points = 51                     # number of grid points
x_delta = L_x / (num_points - 1)    # Spatial descretization
A_init = 1                          # Amplitude = 1
lambda_init = 50                    # Iniital wavelength = 50m



def FTBS(t_delta,lambda_init):                              #Input values of Lambda to get back value of x_delta and t_delta for time step
  t_steps = int(t_total/t_delta)                                 #Time descritezation
  courant = v*t_delta/x_delta                                    #Calculate Courant number
  x_mesh = np.linspace(min_x,max_x,num_points)                   #Spatial descritization
  U_i = np.zeros((t_steps+1,num_points))                         #Creating 2d array where each time step and value of x is descretized
  U_i[0,:] = A_init*np.sin(2*np.pi*x_mesh/lambda_init)           #Setting the initial boundary condition at t = 0
  # U_i[0,:] =np.sign(A_init*np.sin(2*np.pi*x_mesh/lambda_init))   #Boundary condition for sign function
  U_exact = U_i[0,:]                                             #Boundary condition t = 0 is going to be equal to U_exact
  U_error = np.zeros(t_steps+1)                                  #Creating numpy array for value of error over time
  U_max = np.zeros(t_steps+1)                                    #Creating numpy array for calculating

  for n in range (0,t_steps):

    for i in range(1,num_points):
      U_i[n+1,i] = U_i[n,i]-courant*(U_i[n,i]-U_i[n,i-1])
    U_i[n+1,0] = U_i[n+1,num_points-1]

    U_exact = A_init * np.sin(2*np.pi*(x_mesh-v*(n+1)*t_delta)/lambda_init)
    error = abs(U_exact-U_i[n+1])
    max_error = np.max(error)
    max_u = np.max(U_i[n+1])
    U_error[n] = max_error
    U_max[n] = max_u
    
    if((n%10 == 0)): 
      plt.clf()
      plt.scatter(x_mesh,U_i[n+1,:],marker = 'o',color = 'r',label = "U numerical solution")
      plt.plot(x_mesh,U_exact,color = 'b',label = "U exact solution")
      plt.title("U numerical vs U exact with delta t = {} and C = {} @ time = {} ".format(t_delta,courant,t_delta*(n+1)))
      plt.xlabel("x")
      plt.ylabel("U")
      plt.legend()
      plt.draw();plt.pause(0.0001)
      plt.show()

  plt.title("Max error of U numerical vs U exact ")
  plt.plot(np.arange(0,t_steps+1),U_error,color = 'r',label = "Max Error")
  plt.xlabel('Time')
  plt.ylabel('Error')
  plt.xscale('log')
  plt.yscale('log')
  plt.legend()
  plt.show()

  plt.title("Max Amplitude of U numerical")
  plt.plot(np.arange(0,t_steps+1),U_max,color = 'r',label = "Max Error")
  plt.xlabel('Time')
  plt.ylabel('Amplitude')
  plt.yscale('log')
  plt.legend()
  plt.show()

FTBS(2,50)      # Experiment 1
FTBS(1,50)      # Experiment 2
FTBS(0.5,50)    # Experiment 2
FTBS(4,50)      # Experiment 3
FTBS(8,50)      # Experiment 3
FTBS(2,10)      # Experiment 4
FTBS(1,10)      # Experiment 4
# FTBS(1,50)      # Experiment 5
```
